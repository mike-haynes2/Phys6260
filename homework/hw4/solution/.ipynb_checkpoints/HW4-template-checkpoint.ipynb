Below I show an example of how to create an animation in an notebook, which you will need for each part in the homework.

```python
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.rcParams['animation.html'] = 'html5' # this is used to display animations in jupyter notebooks
plt.rcParams['animation.ffmpeg_path'] = '/home/chaynes32/.local/lib/python3.10/site-packages/ffmpeg'
```

```python
solutions = []
niter = 20
N = 100
allt = np.linspace(0,1,niter)
x = np.linspace(0,1,N)
for t in allt:
    solutions.append(np.sin(2 * np.pi * (x-t)))
```

```python
ims = []
fig = plt.figure(figsize=[4,3])
plt.grid()

i = 0
plot_every = 1
for sol in solutions:
    if (i % plot_every == 0): # output frequency for frames  
        im = plt.plot(x, sol, 'k-', animated=True)
        plt.ylim(-1,1)
        ims.append(im)
    i+=1
ani = animation.ArtistAnimation(fig, ims, interval=100, blit=True,
                                repeat_delay=1000)
# ani.save('ftcs.mp4')   
ani
```
