## Problem 1
### Adaptive Integration

We use an adaptive trapezoidal method to approximate the value of 
$$ I = \int_0^1 \sin^2{\sqrt{100x}} \, \mathrm{d}x \qquad .$$
Our program will start with a single integration slice and increase the number of slices by a factor of two until reaching the desired error of $\epsilon = 10^{-6}$. 

After each iteration $i$, we will output the estimate of $I_i$, $\epsilon_i$, and $N_i=2^{i}$ (for $i\in\{0,1,2,..,T\}$, where $T$ is the number of iterations required).


```python
import numpy as np
import math as m
# Func to calculate each successive iteration of trapezoid rule,
# using the adaptive method derived in Lecture notes 03.5.3

def func(x):
    return ( (np.sin(np.sqrt(100.*x))) ** 2 )


def AdaptTrap(f, N, I_prev, h, a, b):
    c = I_prev / 2.
    s = 0
    for k in range(1,N):
        if k % 2 == 0:
            continue
        s += h*f(a + k*h)
    I_i = c + s
    I_err = (I_i - I_prev)/3.
    return I_i, I_prev, I_err



a = 0.
b = 1.
N = 1
eps = 1.e-06
h = 0.

Int_i = 0.
Int_im1 = 0.
err_i = 1.

# A trapezoid rule approximation with 1 slice is given by
Trap0 = (func(b) - func(a))/2.

Int_i = Trap0
# Iterate the approximation
it = 1
while np.abs(err_i) > eps:
    N = 2 ** it
    h = (b-a)/N
    Int_i, Int_im1, err_i = AdaptTrap(func, N, Int_i, h, a, b)
    print("Iteration no. "+str(it)
          +": I_"+str(it)+" = "+str(round(Int_i,7))
          +", error = "+str(round(err_i,7)))
    it += 1

print("Adaptive integration took "+str(it-1)+" iterations to converge to the provided tolerance: "+str(eps))


```

### Heat Capacity of a Solid

Using the expression 
$$ C_V = 9k_B V n  \left(\frac{T}{\theta_D}\right)^3 \int_0^{\theta_D/T} \frac{x^4 e^x}{(e^x - 1)^2} \, dx \qquad ,$$
we can calculate $C_V$ for a given temperature (here $n$ is number density, and we will use $\rho := mn$ as _mass_ density). For the material specified, i.e., $(n=6.022 \cdot 10^{28}\,\mathrm{m}^{-3}, \theta_D = 428 \,\mathrm{K}$), we will use Gaussian quadrature to evaluate the integral in the expression above.

Note that we will define the variable $\tau \equiv T/\theta_D$, such that the above expression can be rewritten:
$$ C_V = 9k_B V n \tau^3 \int_0^{\tau^{-1}} \frac{x^4 e^x}{(e^x - 1)^2} \, dx \qquad ,$$

### Part (a)

```python
from scipy import constants
# Setup Gaussian Quadrature operation
from gaussxw import gaussxw

def gQuad(N, a, b, f):
    xg, wg = gaussxw(N)
    xp = 0.5*(b-a)*xg + 0.5*(b+a)
    wp = 0.5*(b-a)*wg
    gq = (wp * f(xp)).sum()
    return gq

def Debye_Integrand(x):
    num = (x ** 4.) * np.exp(x)
    den = (np.exp(x) - 1.) ** 2.
    return (num/den)


def cv(tau):
    const = 9.*constants.k
    # define material parameters (in SI units)
    n = 6.022e+28
    V = 0.001 # (1000 cubic centimeters)
    front = const * V * n * (tau ** 3.)
    # initialize integration limits
    ulim = 1. / tau
    llim = 0.
    I_debye = gQuad(50,llim,ulim,Debye_Integrand)
    val = front * I_debye
    return val

# theta_D = 428.
# tau_test = 5/theta_D

    
```

## Problem 3
### Wave Diffraction

Using the expression 
$$ \frac{I}{I_0} = \frac{1}{8}\left( \left(2C(u) + 1\right)^2 + \left(2S(u) + 1\right)^2 \right) \qquad ,$$
we can determine the intensity of the diffracted wave at a point $(x,z)$ past the edge according to near-field diffraction theory. Here, $I_0$ is the undiffracted intensity, $$u=x\sqrt{\frac{2}{\lambda z}} \, , \qquad C(u) = \int_0^u \cos{(\pi t^2 / 2)} \, \mathrm{d}t \, , \qquad S(u) = \int_0^u \sin{(\pi t^2 / 2)} \, \mathrm{d}t \qquad .$$
We need to calculate the quantity $I/I_0$ for $x\in [-5\,\mathrm{m},5\,\mathrm{m} ]$ using the above expression and the Gaussian Quadrature method (for $N=50$). 

```python
# Setup Gaussian Quadrature operation
from gaussxw import gaussxw


def gQuad(N, a, b, f):
    xg, wg = gaussxw(N)
    xp = 0.5*(b-a)*xg + 0.5*(b+a)
    wp = 0.5*(b-a)*wg
    gq = (wp * f(xp)).sum()
    return gq

## define I/I_0 (C_u, S_u) in terms of the integral functions (treated here as variables
def calcI_I0(Cu,Su):
    p1 = 2. * Cu + 1.
    p2 = 2. * Su + 1.
    ret = (1./8.) * ( (p1 ** 2.) + (p2 ** 2.) )
    return ret

## define C(u), S(u) function integrands. define u(x,z,\lambda)

def C(t):
    return np.cos(m.pi* t ** 2. / 2.)

def S(t):
    return np.sin(m.pi* t ** 2. / 2.)

def calcu(x,z,lam):
    return (x*m.sqrt(2./(lam * z)))


N = 50
a = 0.
# wavelength \lambda set to 1 m
lam = 1.
# z set to 3 meters
z = 3.
# set stepsize
dx = 0.05



x_arr = np.arange(-5.,5.,dx)
I_I0_arr = np.zeros_like(x_arr)

x_list = x_arr.tolist()

for i, x in enumerate(x_list):
    # assign u coordinate using x,z and wavelength
    u = calcu(x,z,lam)
    # set upper limit of integration to u
    b = u
    # evaluate integrals
    C_u = gQuad(N,a,b,C)
    S_u = gQuad(N,a,b,S)
    I_I0_arr[i] = calcI_I0(C_u,S_u)
    
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6,6))

plt.plot(x_arr, I_I0_arr, linewidth=2.,color='navy')
ax.set_xlabel(r'$x$ [m]')
ax.set_ylabel(r'$\frac{I}{I_0}$', fontsize=16)
fig.suptitle(r"Relative Intensity $I/I_0$", fontsize=24)
ax.grid()
plt.savefig('Intensity_plot_hw2.3.png',dpi=275)
```

### Lotka-Volterra Equations

The Lotka-Volterra model for predator-prey interactions describes the relationship between the two populations with the following system of ordinary differential equations (dependent on time):
\begin{align}
%prey population x
\frac{\mathrm{d}x}{\mathrm{d}t} = \alpha x - \beta x y \qquad ,\\
%predator population y 
\frac{\mathrm{d}y}{\mathrm{d}t} = \gamma x y - \delta y  \,\,\qquad .
\end{align}
Here, $x$ represents the populations of the prey (rabbits), and $y$ the population of the predators (foxes). We will simulate this population for some time $t\longrightarrow30$.

### (a)

```python
import numpy as np
# set predefined constants
alpha = 1.
beta = 0.5
gamma = 0.5
delta = 2.

h = 0.1

t_max = 30

x0 = y0 = 2.

## Define functions for dx/dt and dy/dt (= F_y)
def F_x(alpha, beta, x, y, t):
    c1 = alpha * x
    c2 = -beta * x * y
    return (c1 + c2)

def F_y(gamma,delta, x, y, t):
    c1 = gamma * x * y
    c2 = -delta * y
    return (c1 + c2)

def F_r(alpha,beta,gamma,delta, x, y, t):
    Fx = F_x(alpha, beta, x, y, t)
    Fy = F_y(gamma,delta, x, y, t)
    return Fx,Fy

# initialization
tA = np.arange(start=0,stop=t_max,step=h)
xA =np.zeros_like(tA)
yA =np.zeros_like(tA)
xA[0] = x0
yA[0] = y0

rA = np.array([xA,yA])
tA_list = tA.tolist()


# execute RK4 routine (vectorized as described by equations (27-34) in section 8.2 of 04 notes
for n,tcur in enumerate(tA_list):
    if n == 0:
        continue
    # following the approach from class notes 
    xcur = rA[0,n-1]
    ycur = rA[1,n-1]
    kc1 = h * np.array(F_r(alpha,beta,gamma,delta, xcur, ycur, tcur))
    kc2 = h * np.array(F_r(alpha,beta,gamma,delta, xcur + 0.5 *kc1[0], ycur + 0.5 *kc1[1], tcur + 0.5*h) )
    kc3 = h * np.array(F_r(alpha,beta,gamma,delta, xcur + 0.5 *kc2[0], ycur + 0.5 *kc2[1], tcur + 0.5*h) )
    kc4 = h * np.array(F_r(alpha,beta,gamma,delta, xcur + kc3[0], ycur + kc3[1], tcur + 0.5*h) )
    rA[:,n] = rA[:,n-1] + (kc1 + 2*kc2 + 2*kc3 + kc4) * (1./ 6.)


#plot the output of vectorized RK4 implementation. 
# note that the constants (alpha,beta,gamma,delta) are given without dimensions specified, so the unit of time is arbitrary (with no dimension)

fig, ax = plt.subplots(figsize=(8,6))

plt.plot(tA,rA[0,:]*1.e+03,color='grey',label='Rabbit population $x(t)$', linewidth=1.8)
plt.plot(tA,rA[1,:]*1.e+03,color='firebrick',label='Fox population $y(t)$', linewidth=1.8)
plt.grid()
plt.legend()
ax.set_xlabel('time [unspecified units]')
ax.set_ylabel(r'population $N$')
fig.suptitle('Population evolution of Predator-Prey system', fontsize=18)
#plt.savefig('p4_fig1.png',dpi=290)
plt.show()
```

### (b)
In the above figure, we show the time evolution of the predator-prey system over 30 (unspecified) time units. Immediately, the cyclical nature of the system is apparent. To understand this periodic variation in population, we consider first a single 'period' of the population dynamics: e.g., from $t=0$ to $t\approx4.6$. Just after the population is initialized (with $x=y=2000$), the rabbit population grows tremendously and the fox population declines. After about 3 time units, the rabbit population begins to taper due to a concurrent positive rate of change in the fox population. The constants $\alpha,\beta,\gamma,\delta$ together configure at what threshold population scale for the prey does the predator population begin to grow again. Moving from $t=3\longrightarrow t=4$, the predator (fox) population grows rapidly before intersecting the population of the prey (rabbit). As the prey population declines, so does the predator population, though shortly out of phase. Between $t=4$ and $t\approx4.6$, both populations decrease to intersect once again at a population number of 2000, equal to the initial condition. Given the stability of this oscillation and how the pattern is resolved even with a large timestep $h=1$, we believe this nonlinear system to exhibit a RPO (relative periodic orbit) in predator-prey phase space when initialized with these conditions.

A friend of mine (fellow GT physics grad) works in Atlanta and one of his recent projects was to model the transport of a laser pulse over a duration of several 100s of meters through the city. A laser is created by a Gaussian cavity, essentially two mirrors which repeatedly reflect (and thus collimate) the photons into coherence. On one end of the cavity, the mirror has a nonzero probability of allowing the transmission of a photon, so the beam itself is emitted as the subset of photons transmitted through this mirror at every point in time. Depending on the material used inside the cavity, the laser will have different wavelengths/colors and different efficiencies. As a beam of coherent photons is emitted by the laser lens, the beam begins to spread. This is described by the beam waist in two orthogonal directions transverse to the direction of propogation. The beam waist(s) decrease for a short period until they reach the focal point, after which they grow as a function of distance travelled. To put this growth into context, consider the following: for typical commercially-available pocket lasers, the beam waist growth is such that if you shine a laser pointer at the Moon, the beam will cover $\approx1/3$ of its surface when it reaches it. 

However, this calculation my friend is carrying out is not just for "fun". They need to position a receiver in the exact location with the proper aperture to receive the signal. On top of these (quasi-)linear propagation effects described above, higher order effects also play a role here. For instance, the propagation of the light is modulated by local atmospheric density, which in turn changes with local pressure gradients. Other optical phenomena (scattering, partial refraction) occur when the beam encounters suspended heavier particlates (i.e., aerosols). Therefore, in a (typically) warm, sunny city that is busy like Atlanta, there are essentially constant gradients in wave propagation along its course.
