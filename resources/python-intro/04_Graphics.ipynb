## First example
* Plotting a list of numbers
* When given a list, matplotlib will assume that they are the `y` values and show them versus a list of integers (0 -> N).
* In notebooks, you show always issue the `show()` routine to display the graph.  Otherwise, the plot might not display correctly.

```python
y = [19, 11, 2, 10.1, 9, 1]
plt.plot(y)
plt.show()
```

* We can modify various aspects of the plot.
* First let's look at the line style and colors.
* Color is denoted by a single character: r,g,b,c,m,y,k,w indicate red, green, blue, cyan, magneta, yellow, black, and white.
* Line styles are denoted by "-", "--", ":", "-." for solid, dashed, dotted, and dash-dotted lines
* Data points can be shown with "s", "o", "^" for square, circle, and triangle points
* There are many different options, which can be found at http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot

```python
plt.plot(y, "r:")  # c = color, ls = linestyle
plt.show()
```

```python
plt.plot(y, c="g", ls=":", marker="s", mec="b", mew=3, ms=20, lw=4)
plt.show()
```

```python
plt.plot(y, "g:s")#, marker="s")
#plt.plot(y, c="g", ls=":", marker="s")
plt.show()
```

```python
plt.plot(y, c='g', ls=':', marker='s', mec='r', ms=16, mew=3, lw=3)  # ms = marker size, mec = marker edge color, mew = marker edge width, lw = linewidth
plt.show()
```

### Including x values
* It's straightforward to give corresponding x-values to the y-values
* They need to have the same length.

```python
x = [1,1.2,2,3.6,5,10]
plt.plot(x, y, c='g', lw=3, marker='o')  # lw = linewidth
plt.show()
```

Let's add some axis labels and fine tune the axis ranges.

```python
plt.xlabel('x [m]')
plt.ylabel('y [N]')
plt.xlim(0.9, 10.1)
plt.plot(x, y, c='g', lw=3, marker='o')  # lw = linewidth
plt.show()
```

### Two plots with functions together
* Let's plot two lines in a figure
* We'll use `sin()` and `cos()` as example.
* First create a `theta` variable that goes from 0 to $4\pi$ and then plot them.
* One calls several `plot()` instances to put the lines on the same figure.  Then finalize the image with `show()`.
* Strings can contain $\LaTeX$ but needs to start with an `r` to tell matplotlib that it's a "raw" string

```python
import numpy as np
theta = np.linspace(0, 4*np.pi, 100)  # 0 to 4pi with 100 elements
plt.plot(theta, np.sin(theta), c='k', ls=':', lw=3, label=r'sin($\theta$)')  # label is for the legend
plt.plot(theta, np.cos(theta), c='g', ls='-', lw=3, label=r'cos($\theta$)')
plt.xlabel(r'$\theta$')
plt.ylabel('Value')
plt.legend(loc='lower left')  # loc is where the legend is placed.  "best" usually does a good job
plt.show()
```
