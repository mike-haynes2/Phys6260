```python
from numpy.random import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize'] = (5,4)
matplotlib.rcParams['font.size'] = 18
```

```python
niter = 20
pi_results = np.zeros(niter-1)
for it in range(1,niter):
    N = 2**it
    x = random(N)
    y = random(N)
    r = x**2 + y**2
    pi_results[it-1] = 4 * (r < 1).sum() / N
```

```python
plt.semilogx(2**np.arange(1,niter), pi_results)
plt.axhline(np.pi, ls='--', c='k')
plt.xlabel('Number of samples')
plt.ylabel(r'Estimate of $\pi$')
plt.show()
```

```python
xx = 2**np.arange(1,niter)
yy = np.abs(pi_results-np.pi)/np.pi
plt.loglog(xx, yy)
plt.loglog(xx, yy[0] * np.sqrt(xx[0]/xx), ls='--')
plt.xlabel('Number of samples')
plt.ylabel(r'Error estimtae of $\pi$')
plt.show()
```
