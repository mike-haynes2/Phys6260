### RK4 -- pendulum equation
#### Using script in class (virtual meeting 01/15)

```python
#%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

g = 9.81  # gravity
l = 0.25   # arm length of pendulum [m]
theta0 = 10.0 # initial arm position [degrees]

freq = np.sqrt(g/l)
period = 2.* np.pi * (1./freq)

def f(r,t):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = -(g/l) * np.sin(theta)
    return np.array([ftheta,fomega],float)

a = 0.0
b = 10.0
N = 1000
h = (b-a)/N

tpoints = np.arange(a,b,h)
r = np.zeros((N,2))
r[0] = [np.pi * theta0/180.0, 0.0]

# Loop over time to solve for x and y, which will be stored in r[]
for i,t in enumerate(tpoints[1:],1):
    k1 = h * f(r[i-1], t)
    k2 = h * f(r[i-1]+0.5*k1, t+0.5*h)
    k3 = h * f(r[i-1]+0.5*k2, t+0.5*h)
    k4 = h * f(r[i-1]+k3, t+h)
    r[i] = r[i-1] + (k1 + 2*k2 + 2*k3 + k4) / 6.0

plt.plot(tpoints, r[:,0], label=r'$\theta$')
plt.plot(tpoints, r[:,1], label=r'$\omega$')
plt.xlabel("t")
plt.ylabel(r"$\theta, \omega$")
plt.legend()
plt.show()
print("Oscillatory properties predicted by SHM: \nPeriod [s]: "+str(period))
```
